```python
#por si acaso, creo copia y trabajo sobre ella:
df1 = df.copy()
```

```python
#modificamos todo a minúsculas. Pongo un rename para que se queden los cambios en la copi del df:
df1.columnas = df1.rename(columns=str.lower,inplace=True)
df1.columnas
```

```python
df1.head()
```

```python
#reemplazamos coma por punto y pasamos a float:
def modificacion_puntuacion(df1, columnas):
    for columna in columnas:
        df1[columna] = df1[columna].str.replace("," , ".").astype(float)
        
mod_columnas = ['danceability', 'energy', 'key', 'loudness', 'speechiness', 'acousticness', 'instrumentalness']

modificacion_puntuacion(df1, mod_columnas)
df1.head()
```

```python
#verificamos que sean del tipo float. Como lo especifiqué en la función, debe salir todo correcto:
df1.dtypes
```

```python
#reemplazamos las columnas track, title y album a todo minúsculas:
def titulo_a_minuscula (df1,columnas):
    for columna in columnas:
        df1[columna] = df1[columna].str.lower()

titulo_a_minuscula(df1,["track", "title", "album"])

df1.head()
```

```python
#reemplazamos coma por punto y pasamos a float:
df1["artist"] = df1["artist"].str.replace("$", "s")

df1.head()
```

4. Filtrado de datos:

    - ¿Cuáles son los valores únicos de la columna "album_type"? Crea tres dataframes diferentes, uno para cada tipo de "album_type". ¿Cuántas canciones tenemos en cada tipo?
    
    - Usando el dataframe de los albumes que hemos creado en el ejercicio anterior. ¿Cuál es la media y la desviación estándar de "danceability", "acousticness" y "speechiness" de cada artista?. Muesrtra los resultados en tres dataframes diferentes, uno para cada métrica ("danceability", "acousticness" y "speechiness"). Ordena los resultados de mayor a menor en base a la media.
    
    - Se quiere estudiar si existe una relación entre la cantidad de visitas que reciben los videos de música de un artista en una plataforma de streaming y la cantidad de likes que estos videos obtienen. Por lo tanto, queremos contestar a la siguiente pregunta: ¿Son aquellos artistas que tienen más visitas los que más likes tienen? Para solucionar este ejercicio deberéis:

        - Agrupa por artista y calcular la media de visitas y de likes.
        - Quedate con los 10 artistas que más visitas han tenido.
        - Quedate con los 10 artistas que más likes han tenido.
        - BONUS Haced lo mismo para los datos de tipo "single".

```python
#valores únicos de la columna album_type:
df1['album_type'].unique()
```

```python
#creamos 3 dataframes:
album_df = df1[df1["album_type"] == "album"]
single_df = df1[df1["album_type"] == "single"]
compilation_df = df1[df1["album_type"] == "compilation"]

#visualicemos número de canciones en cada df:
print(album_df.shape, single_df.shape, compilation_df.shape)
```

```python
#media y la desviación estándar de "danceability", "acousticness" y "speechiness" de cada artista dentro de df de album_df:
danceability_desv = album_df.groupby("artist")["danceability"].agg(["mean", "std"]).sort_values(by="mean", ascending=False)
acousticness_desv = album_df.groupby("artist")["acousticness"].agg(["mean", "std"]).sort_values(by="mean", ascending=False)
speechiness_desv = album_df.groupby("artist")["speechiness"].agg(["mean", "std"]).sort_values(by="mean", ascending=False)
```

```python
#agrupamos por artista y calculamos la media de visitas y likes:
visitas_likes_artistas = df1.groupby("artist")[["views", "likes"]].mean()
visitas_likes_artistas
```

```python
# 10 artistas que más visitas y más likes tienen:
top_10_visitas = visitas_likes_artistas.sort_values(by='views', ascending=False).head(10)
top_10_likes = visitas_likes_artistas.sort_values(by='likes', ascending=False).head(10)

#sort_values(): ordeno df por las columnas "views" o "likes".
#ascending=False: ordeno de mayor a menor.
#head(10): cojo las primeras 10 filas después de ordenar.
```
